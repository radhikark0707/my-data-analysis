```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_excel('/content/Superstore.xls')
```

```python
data.head()
```

```python
#Barplot
plt.bar(data['Category'],data['Sales'])
plt.title('Total Sales by category')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.show()
```

```python
#Histogram
plt.title('Distribution of sales')
sns.histplot(data['Sales'])
plt.xlabel('Sales')
plt.show()

```

```python
#Heatmap
selectedc=data.select_dtypes(include=['float64','int64']).columns
plt.title('Correlation matrix')
sns.heatmap(data[selectedc].corr(), annot=True, cmap='coolwarm')
plt.show()
```

```python
#Boxplot
sns.boxplot(data)
plt.show()
```

```python
#Scatter plot
plt.title('Sales vs Profit')
plt.xlabel('Sales')
plt.ylabel('Profit')
sns.scatterplot(data=data, x='Sales',y='Profit')
plt.show()
```
