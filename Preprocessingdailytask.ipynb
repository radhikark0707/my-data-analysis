```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('/content/train_loan_preprocess.csv')
```

```python
data
```

```python
data.dtypes
```

```python
data.isnull().sum()
```

```python
freqgraph=data.select_dtypes(include=['float'])
freqgraph.hist(figsize=(20,15))
plt.show()
```

```python
#filling missing value with median
```

```python
for i in['CoapplicantIncome','LoanAmount','Loan_Amount_Term','Credit_History']:
  data[i]=data[i].fillna(data[i].median())
```

```python
data.isnull().sum()
```

```python
for i in['Married','Gender','Dependents','Self_Employed']:
  data[i]=data[i].fillna(data[i].mode()[0])
```

```python
data.isnull().sum()
```

```python
data.dtypes
```

```python
for i in ['ApplicantIncome','CoapplicantIncome','LoanAmount','Loan_Amount_Term','Credit_History']:
  plt.figure()
  plt.boxplot(data[i])
  plt.title(i)
```

```python
for i in ['ApplicantIncome','CoapplicantIncome','Loan_Amount_Term','Credit_History']:
  q1=data[i].quantile(0.25)
  q2=data[i].quantile(0.50)
  q3=data[i].quantile(0.75)
  iqr=q3-q1
  lb=q1-1.5*iqr
  ub=q3+1.5*iqr
  outlier=data[(data[i]<lb) | (data[i]>ub)].index
  sns.boxplot(y=outlier)
  plt.title(i)
  plt.show()
  print(i,":","q1=",q1,"q2=",q2,"q3=",q3,"iqr=",iqr,"lb=",lb,"ub=",ub,"Outlier=",outlier)
```

```python
#Label encoding for binary variables
```

```python
data3=data.copy()
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
data3['Gender']=le.fit_transform(data3['Gender'])
data3['Married']=le.fit_transform(data3['Married'])
data3['Self_Employed']=le.fit_transform(data3['Self_Employed'])
data3['Loan_Status']=le.fit_transform(data3['Loan_Status'])
```
