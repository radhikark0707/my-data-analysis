```python
#For numerical column of your choice , calculate mean , median and standard deviation
mychoice='Legislative District'
mean=df[mychoice].mean()
median=df[mychoice].median()
standarddeviation=df[mychoice].std()
print(f' mean= {mean}, median={median}, standard deviation={standarddeviation}')
```

```python
#Identify any potential outliers in a numerical column of your choice.Explain your approach
q1=df['Electric Range'].quantile(0.25)
q3=df['Electric Range'].quantile(0.75)
iqr=q3-q1
lb=q1-1.5*iqr
ub=q3+1.5*iqr
print(f'q1={q1} , q3={q3}, iqr={iqr},lower bound={lb},Upper bound={ub}')
outlier=df[(df['Electric Range']<lb) | (df['Electric Range']>ub)]['Electric Range']
print(outlier)
```

```python
#Find correlation between numerical columns and discuss any intersting relationship
c=['Model Year','Electric Range','Base MSRP','DOL Vehicle ID']
df[c].corr()

```
