```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Apply mapping
data['Select Your Internship Program']=data['Select Your Internship Program'].replace(internship_mapping)
```

Average Rating For Each Aspect Per Day(BARCHART)

```python
data.columns
```

```python
data['Select the Grooming Session Day'].unique()
```

```python
#above column is categorical with ordered days
day_order=['Day 1','Day 2','Day 3','Day 4','Day 5']
data['Select the Grooming Session Day']=pd.Categorical(data['Select the Grooming Session Day'],categories=day_order,ordered=True)
```

```python
aspect=['Todays grooming session was relevant and helpful for me to relate',
                 'Todays delivery of the content was clear and understandable',
                 'How would you rate todays instructor?',
                 'How would you rate todays coordinator support?']
```

```python
#group by select the grooming session day and calculate the avg rating for above aspect
average_rating_day=data.groupby('Select the Grooming Session Day',observed=False)[aspect].mean()
```

```python
fig,axs=plt.subplots(2,2,figsize=(15,10))
average_rating_day['Todays grooming session was relevant and helpful for me to relate'].plot(kind='bar',ax=axs[0,0])
axs[0,0].set_title('Todays grooming session was relevant and helpful for me to relate')
axs[0,0].set_xlabel('Day')
axs[0,0].set_ylabel('Average Rating')

average_rating_day['Todays delivery of the content was clear and understandable'].plot(kind='bar',ax=axs[0,1])
axs[0,1].set_title('Todays delivery of the content was clear and understandable')
axs[0,1].set_xlabel('Day')
axs[0,1].set_ylabel('Average Rating')

average_rating_day['How would you rate todays instructor?'].plot(kind='bar',ax=axs[1,0])
axs[1,0].set_title('How would you rate todays instructor?')
axs[1,0].set_xlabel('Day')
axs[1,0].set_ylabel('Average Rating')

average_rating_day['How would you rate todays coordinator support?'].plot(kind='bar',ax=axs[1,1])
axs[1,1].set_title('How would you rate todays coordinator support?')
axs[1,1].set_xlabel('Day')
axs[1,1].set_ylabel('Average Rating')

plt.tight_layout()
plt.show()
```

Common challenge faced by students

```python
cc=data['Any additional comments/suggestions'].value_counts()
```

```python
cc
```

```python
from wordcloud import WordCloud
import re
```

```python
def preprocess_comments(comments):
  comments=comments.str.lower()  #convert to lowercase
  comments=comments.apply(lambda x:re.sub(r'[^a-z\s]','',x)) #remove punctuation and no
  return comments
```

```python
comments=preprocess_comments(data['Any additional comments/suggestions'])
```

```python
all_comments=' '.join(comments.dropna()) #join all into single string
```

```python
wordcloud=WordCloud(width=800,height=400,background_color='white',stopwords=(['no','nothing','none','na','nil','n/a','nope',
                                                                             'this','is','was','and','for','to','in','of' ])).generate(all_comments)
```

```python
plt.figure(figsize=(10,5))
plt.imshow(wordcloud,interpolation='bilinear')
plt.axis('off')
plt.title('Common challenge faced by students')
plt.show()
```

Trend of average ratings over time

```python
plt.figure(figsize=(10,5))
for column in columns_average:
  plt.plot(daily_avg_rating['Date'],daily_avg_rating[column],label=column)
plt.title('Trend of average ratings over time')
plt.xlabel('Date')
plt.ylabel('Average Rating')
plt.legend()
plt.grid(True)
plt.show()
```

Trend by Internship program

```python
internship_trend=data.groupby(['Date','Select Your Internship Program'])[columns_average].mean().reset_index()
```

```python
for i in columns_average:
  plt.figure(figsize=(10,5))
  for j in data['Select Your Internship Program'].unique():
    program_data=internship_trend[internship_trend['Select Your Internship Program']==j]
    plt.plot(program_data['Date'],program_data[i],label=j)
  plt.title(f'Trend of {i} by Internship program')
  plt.xlabel('Date')
  plt.ylabel('Average Rating')
  plt.legend()
  plt.grid(True)
  plt.show()
```
