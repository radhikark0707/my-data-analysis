Create a new column 'income_binary' where <=50K is 0 and >50K is 1.

```python
def func(income):
  if income== '<=50K':
    return 0
  elif income== '>50K':
    return 1
  else:
     return None
auto_data['income_binary']=auto_data['income'].apply(lambda x:func(x))
auto_data

```

Create a new column 'age_group' categorizing ages into 'Youth' (< 30), 'Adult' (30-60), and 'Senior' (60+).

```python
def funage(age):
  if age < 30:
    return 'Youth'
  elif age>=30 & age <= 60 :
    return 'Adult'
  elif age >60:
    return 'Senior'
  else:
    return None
auto_data['age_group']=auto_data['age'].apply(lambda x:funage(x))
auto_data
```

Replace the missing values represented by '?' in 'workclass' and 'occupation' columns with 'Unknown'.

```python
auto_data['workclass'].replace('?','Unknown',inplace=True)
auto_data['occupation'].replace('?','Unknown',inplace=True)
auto_data
```
